# Redes Neuronales

Las redes neuronales artificiales contienen neuronas artificiales que se denominan unidades. Estas unidades están dispuestas en una serie de capas que juntas constituyen toda la Red Neuronal Artificial de un sistema. Una capa puede tener solo una docena de unidades o millones de unidades, ya que esto depende de cómo se requerirán las redes neuronales complejas para aprender los patrones ocultos en el conjunto de datos. Comúnmente, la red neuronal artificial tiene una capa de entrada, una capa de salida y capas ocultas. La capa de entrada recibe datos del mundo exterior que la red neuronal necesita analizar o aprender. A continuación, estos datos pasan a través de una o varias capas ocultas que transforman la entrada en datos valiosos para la capa de salida. Por último, la capa de salida proporciona una salida en forma de respuesta de las redes neuronales artificiales a los datos de entrada proporcionados.

En la mayoría de las redes neuronales, las unidades están interconectadas de una capa a otra. Cada una de estas conexiones tiene pesos que determinan la influencia de una unidad sobre otra unidad. A medida que los datos se transfieren de una unidad a otra, la red neuronal aprende más y más sobre los datos, lo que finalmente da como resultado una salida de la capa de salida.

Las redes neuronales se dividen en varios tipos:

1. **Red neuronal feedforward:** La red neuronal feedforward es una de las redes neuronales artificiales más básicas. En esta ANN, los datos o la entrada proporcionada viajan en una sola dirección. Entra en la RNA a través de la capa de entrada y sale a través de la capa de salida, mientras que las capas ocultas pueden o no existir. Por lo tanto, la red neuronal feedforward solo tiene una onda propagada hacia adelante y, por lo general, no tiene propagación hacia atrás.

2. **Red neuronal convolucional:** Una red neuronal convolucional tiene algunas similitudes con la red neuronal feed-forward, donde las conexiones entre unidades tienen pesos que determinan la influencia de una unidad sobre otra unidad. Pero una CNN tiene una o más capas convolucionales que utilizan una operación de convolución en la entrada y luego pasan el resultado obtenido en forma de salida a la siguiente capa. CNN tiene aplicaciones en el procesamiento de voz e imágenes, lo que es particularmente útil en la visión por computadora.

3. **Red neuronal recurrente:** la red neuronal recurrente guarda la salida de una capa y devuelve esta salida a la entrada para predecir mejor el resultado de la capa. La primera capa de la RNN es bastante similar a la red neuronal feed-forward y la red neuronal recurrente comienza una vez que se calcula la salida de la primera capa. Después de esta capa, cada unidad recordará alguna información del paso anterior para que pueda actuar como una celda de memoria en la realización de cálculos.

4. **Red Neuronal Modular:** Una red neuronal modular contiene una colección de diferentes redes neuronales que funcionan de forma independiente para obtener la salida sin interacción entre ellas. Cada una de las diferentes redes neuronales realiza una subtarea diferente al obtener entradas únicas en comparación con otras redes. La ventaja de esta red neuronal modular es que descompone un proceso computacional grande y complejo en componentes más pequeños, disminuyendo así su complejidad sin dejar de obtener el resultado requerido.

## Redes neuronales convolucionales

Una **Red Neuronal Convolucional (CNN)** es un tipo de arquitectura de red neuronal de aprendizaje profundo que se usa comúnmente en Computer Vision. La visión artificial es un campo de la Inteligencia Artificial que permite a un ordenador comprender e interpretar la imagen o los datos visuales.

Cuando se trada de aprendizaje automático, las redes neuronales artificiales funcionan muy bien. Las redes neuronales se utilizan en varios conjuntos de datos como imágenes, audio y texto. Diferentes tipos de Redes Neuronales se utilizan para diferentes propositos, por ejemplo para predecir la secuencia de palabras se usan **Redes Neuronales Recurrentes** más precisamente un LSTM, de igual manera para la clasificación de imágenes utilizamos Redes Neuronales de Convolución.

En una Red Neuronal normal hay tres tipos de capas:

1. **Capas de entrada:** Es la capa en la que damos entrada a nuestro modelo. El número de neuronas en esta capa es igual al número total de características en nuestros datos (número de píxeles en el caso de una imagen).

2. **Capa oculta:** La entrada de la capa de entrada se introduce en la capa oculta. Puede haber muchas capas ocultas dependiendo de nuestro modelo y tamaño de datos. Cada capa oculta puede tener un número diferente de neuronas, que generalmente son mayores que el número de entidades. La salida de cada capa se calcula multiplicando la matriz de la salida de la capa anterior con los pesos aprendibles de esa capa y luego mediante la adición de sesgos aprendibles seguidos de la función de activación que hace que la red no sea lineal.

3. **Capa de salida:** La salida de la capa oculta se introduce en una función logística como sigmoid o softmax que convierte la salida de cada clase en la puntuación de probabilidad de cada clase.


Los datos se introducen en el modelo y la salida de cada capa se obtiene del paso anterior que se llama **feedforward**, luego calculamos el error usando una función de error, algunas funciones de error comunes son la entropía cruzada, el error de pérdida cuadrada, etc. La función de error mide el rendimiento de la red. Después de eso, retropropagamos en el modelo calculando las derivadas. Este paso se denomina **retropropagación**, que básicamente se utiliza para minimizar la pérdida.



La red neuronal convolucional (CNN) es la versión extendida de las redes neuronales artificiales (ANN) que se utiliza principalmente para extraer la característica del conjunto de datos de matriz en forma de cuadrícula. Por ejemplo, conjuntos de datos visuales como imágenes o vídeos, donde los patrones de datos desempeñan un papel importante.

### Arquitectura de CNN
La red neuronal convolucional consta de varias capas, como la capa de entrada, la capa convolucional, la capa de agrupación y las capas totalmente conectadas.

La capa convolucional aplica filtros a la imagen de entrada para extraer entidades, la capa Agrupación reduce la resolución de la imagen para reducir el cálculo y la capa totalmente conectada realiza la predicción final. La red aprende los filtros óptimos a través de la retropropagación y el descenso de gradiente.



### ¿Cómo funcionan las capas convolucionales?

Las redes neuronales de convolución o covnets son redes neuronales que comparten sus parámetros. Imagina que tienes una imagen. Se puede representar como un cuboide que tiene su longitud, anchura (dimensión de la imagen) y altura (es decir, el canal, ya que las imágenes generalmente tienen canales rojo, verde y azul).

![](https://media.geeksforgeeks.org/wp-content/uploads/cnn-2-300x133.jpg)


Ahora imagina tomar un pequeño parche de esta imagen y ejecutar una pequeña red neuronal, llamada filtro o kernel en ella, con, digamos, K salidas y representarlas verticalmente. Ahora desliza esa red neuronal a través de toda la imagen, como resultado, obtendremos otra imagen con diferentes anchos, alturas y profundidades. En lugar de solo los canales R, G y B, ahora tenemos más canales pero menos ancho y alto. Esta operación se denomina convolución. Si el tamaño del parche es el mismo que el de la imagen, será una red neuronal normal. Debido a este pequeño parche, tenemos menos pesos.

![](https://media.geeksforgeeks.org/wp-content/uploads/Screenshot-from-2017-08-15-13-55-59-300x217.png)


**Algunos aspectos clave**

* Las capas de convolución consisten en un conjunto de filtros aprendibles (o núcleos) que tienen anchos y alturas pequeños y la misma profundidad que la del volumen de entrada (3 si la capa de entrada es entrada de imagen).

* Por ejemplo, si tenemos que ejecutar convolución en una imagen con dimensiones 34x34x3. El tamaño posible de los filtros puede ser $a\times a\times 3$, donde $a$ puede ser algo así como 3, 5 o 7, pero más pequeño en comparación con la dimensión de la imagen.

* Durante el paso hacia adelante, deslizamos cada filtro a través de todo el volumen de entrada paso a paso, donde cada paso se llama **stride** (que puede tener un valor de 2, 3 o incluso 4 para imágenes de alta dimensión) y calculamos el producto de puntos entre los pesos del kernel y el parche a partir del volumen de entrada.

* A medida que deslizamos nuestros filtros, obtendremos una salida 2D para cada filtro y los apilaremos, como resultado, obtendremos un volumen de salida que tiene una profundidad igual al número de filtros. La red aprenderá todos los filtros.

![](https://www.codificandobits.com/img/posts/2019-03-30/ejemplo-convolucion-imagen-tres-dimensiones.gif)


### Capas utilizadas para crear ConvNets

Una arquitectura completa de redes neuronales de convolución también se conoce como covnets. Un covnets es una secuencia de capas, y cada capa transforma un volumen en otro a través de una función diferenciable.

**Tipos de capas:** conjuntos de datos Tomemos un ejemplo ejecutando una covnets en una imagen de dimensión 32 x 32 x 3.

* **Capas de entrada:** Es la capa en la que damos entrada a nuestro modelo. En CNN, generalmente, la entrada será una imagen o una secuencia de imágenes. Esta capa contiene la entrada sin procesar de la imagen con un ancho de 32, una altura de 32 y una profundidad de 3.

* **Capas convolucionales:** Esta es la capa que se utiliza para extraer la entidad del dataset de entrada. Aplica un conjunto de filtros aprendibles conocidos como kernels a las imágenes de entrada. Los filtros/kernels son matrices más pequeñas, generalmente de forma 2×2, 3×3 o 5×5. Se desliza sobre los datos de la imagen de entrada y calcula el producto escalar entre el peso del kernel y el parche de imagen de entrada correspondiente. El resultado de esta capa se denomina mapas de entidades publicitarias. Supongamos que usamos un total de 12 filtros para esta capa, obtendremos un volumen de salida de dimensión 32 x 32 x 12.

* **Capa de activación:** Al agregar una función de activación a la salida de la capa anterior, las capas de activación agregan no linealidad a la red. Aplicará una función de activación por elemento a la salida de la capa de convolución. Algunas funciones de activación comunes son RELU: max(0, x), Tanh, Leaky RELU, etc. El volumen permanece sin cambios, por lo que el volumen de salida tendrá unas dimensiones de 32 x 32 x 12.

* **Capa de agrupación:** Esta capa se inserta periódicamente en las redes y su función principal es reducir el tamaño del volumen, lo que hace que el cálculo sea rápido, reduzca la memoria y también evite el sobreajuste. Dos tipos comunes de capas de agrupación son la agrupación máxima y la agrupación media. Si usamos una piscina máxima con filtros de 2 x 2 y zancada 2, el volumen resultante será de dimensión 16x16x12.

![](https://media.geeksforgeeks.org/wp-content/uploads/Screenshot-from-2017-08-15-17-04-02.png)


* **Aplanamiento:** los mapas de entidades resultantes se aplanan en un vector unidimensional después de las capas de convolución y agrupación para que se puedan pasar a una capa completamente vinculada para su categorización o regresión.

* **Capas totalmente conectadas:** Toma la entrada de la capa anterior y calcula la tarea final de clasificación o regresión.

![](https://media.geeksforgeeks.org/wp-content/uploads/Screenshot-from-2017-08-15-17-22-40.png)

* **Capa de salida:** La salida de las capas totalmente conectadas se introduce en una función logística para tareas de clasificación como sigmoid o softmax, que convierte la salida de cada clase en la puntuación de probabilidad de cada clase.

**Ventajas de las redes neuronales convolucionales (CNN):**

1. Son buenas para detectar patrones y características en imágenes, videos y señales de audio.

2. Robusto a la traslación, la rotación y la invarianza de escala.

3. Entrenamiento de extremo a extremo, sin necesidad de extracción manual de características.

4. Puede manejar grandes cantidades de datos y lograr una alta precisión.

**Desventajas de las redes neuronales convolucionales (CNN):**

1. Desde el punto de vista computacional, es costoso de entrenar y requiere mucha memoria.

2. Puede ser propenso al sobreajuste si no se utilizan suficientes datos o una regularización adecuada.

3. Requiere grandes cantidades de datos etiquetados.

4. La interpretabilidad es limitada, es difícil entender lo que la red ha aprendido.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from itertools import product
 
# set the param 
plt.rc('figure', autolayout=True)
plt.rc('image', cmap='magma')
 
# define the kernel
kernel = tf.constant([[-1, -1, -1],
                    [-1,  8, -1],
                    [-1, -1, -1],
                   ])
 
# load the image
image = tf.io.read_file('Ganesh.jpg')
image = tf.io.decode_jpeg(image, channels=1)
image = tf.image.resize(image, size=[300, 300])
 
# plot the image
img = tf.squeeze(image).numpy()
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original Gray Scale image')
plt.show();
 
 
# Reformat
image = tf.image.convert_image_dtype(image, dtype=tf.float32)
image = tf.expand_dims(image, axis=0)
kernel = tf.reshape(kernel, [*kernel.shape, 1, 1])
kernel = tf.cast(kernel, dtype=tf.float32)
 
# convolution layer
conv_fn = tf.nn.conv2d
 
image_filter = conv_fn(
    input=image,
    filters=kernel,
    strides=1, # or (1, 1)
    padding='SAME',
)
 
plt.figure(figsize=(15, 5))
 
# Plot the convolved image
plt.subplot(1, 3, 1)
 
plt.imshow(
    tf.squeeze(image_filter)
)
plt.axis('off')
plt.title('Convolution')
 
# activation layer
relu_fn = tf.nn.relu
# Image detection
image_detect = relu_fn(image_filter)
 
plt.subplot(1, 3, 2)
plt.imshow(
    # Reformat for plotting
    tf.squeeze(image_detect)
)
 
plt.axis('off')
plt.title('Activation')
 
# Pooling layer
pool = tf.nn.pool
image_condense = pool(input=image_detect, 
                             window_shape=(2, 2),
                             pooling_type='MAX',
                             strides=(2, 2),
                             padding='SAME',
                            )
 
plt.subplot(1, 3, 3)
plt.imshow(tf.squeeze(image_condense))
plt.axis('off')
plt.title('Pooling')
plt.show()
```

Ejemplo para clasificar imágenes del conjunto de datos CIFAR-10, que contiene imágenes de 10 clases diferentes, como automóviles, pájaros y gatos.



```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Cargar y preprocesar el conjunto de datos CIFAR-10
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalizar los valores de los píxeles
train_images, test_images = train_images / 255.0, test_images / 255.0

# Crear el modelo de la CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
])

# Añadir capas densas en la parte superior
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

# Compilar el modelo
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Resumen del modelo
model.summary()

# Entrenar el modelo
history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))

# Evaluar el modelo
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)

# Gráfico de entrenamiento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')

plt.show()
```

Este código realiza lo siguiente:

* Carga el conjunto de datos CIFAR-10 y lo divide en conjuntos de entrenamiento y prueba.
* Preprocesa las imágenes normalizando los valores de los píxeles.
* Construye una CNN con tres capas convolucionales y dos capas de pooling, seguidas de capas densas para la clasificación.
* Compila el modelo con el optimizador 'adam' y una función de pérdida adecuada para clasificación multiclase.
* Entrena el modelo con los datos de entrenamiento.
* Evalúa el modelo con los datos de prueba y muestra la precisión del modelo.
* Muestra un gráfico de la precisión del entrenamiento y validación a lo largo de las épocas.


## Red Neuronal Recurrente (RNN)

La red neuronal recurrente (RNN) es un tipo de red neuronal en la que la salida del paso anterior se alimenta como entrada al paso actual. En las redes neuronales tradicionales, todas las entradas y salidas son independientes entre sí, pero en los casos en que se requiere predecir la siguiente palabra de una oración, se requieren las palabras anteriores y, por lo tanto, es necesario recordar las palabras anteriores. Así nació RNN, que resolvió este problema con la ayuda de una capa oculta. La característica principal y más importante de RNN es su **estado oculto**, que recuerda cierta información sobre una secuencia. El estado también se conoce como estado de memoria, ya que recuerda la entrada anterior a la red. Utiliza los mismos parámetros para cada entrada, ya que realiza la misma tarea en todas las entradas o capas ocultas para producir la salida. Esto reduce la complejidad de los parámetros, a diferencia de otras redes neuronales.

![](https://media.geeksforgeeks.org/wp-content/uploads/20230518134831/What-is-Recurrent-Neural-Network.webp)

Hoy en día, se utilizan diferentes técnicas de Machine Learning para manejar diferentes tipos de datos. Uno de los tipos de datos más difíciles de manejar y de pronosticar son los datos secuenciales. Los datos secuenciales son diferentes de otros tipos de datos en el sentido de que, si bien se puede suponer que todas las características de un conjunto de datos típico son independientes del orden, esto no se puede suponer para un conjunto de datos secuencial. Para manejar este tipo de datos, se concibió el concepto de Redes Neuronales Recurrentes. Se diferencia de otras Redes Neuronales Artificiales en su estructura. Mientras que otras redes "viajan" en una dirección lineal durante el proceso de retroalimentación hacia adelante o el proceso de retropropagación, la red recurrente sigue una relación de recurrencia en lugar de un paso de retroalimentación hacia adelante y utiliza la propagación hacia atrás a través del tiempo para aprender.



La red neuronal recurrente consta de varias unidades de función de activación fijas, una para cada paso de tiempo. Cada unidad tiene un estado interno que se denomina estado oculto de la unidad. Este estado oculto significa el conocimiento pasado que la red tiene actualmente en un período de tiempo determinado. Este estado oculto se actualiza en cada paso de tiempo para significar el cambio en el conocimiento de la red sobre el pasado. El estado oculto se actualiza mediante la siguiente relación de periodicidad:

$h_t = fw(x_t, h_{t-1})$

Donde

$h_t$ es el nuevo estado oculto

$h_{t-1}$ Es el estado oculto anterior

$x_t$ Es el input actual

Nota: Por lo general, para comprender los conceptos de una red neuronal recurrente, a menudo se ilustra en su forma desenrollada.

En cada periodo de tiempo, el nuevo estado oculto se calcula utilizando la relación de periodicidad indicada anteriormente. Este nuevo estado oculto generado se utiliza para generar un nuevo estado oculto y así sucesivamente.

El flujo de trabajo básico de una red neuronal recurrente es el siguiente:

![](https://media.geeksforgeeks.org/wp-content/uploads/20190701121614/unrolled.png)

Tenga en cuenta que $h_{0}$   es el estado oculto inicial de la red. Normalmente, es un vector de ceros, pero también puede tener otros valores. Un método consiste en codificar las presunciones sobre los datos en el estado oculto inicial de la red. Por ejemplo, para que un problema determine el tono de un discurso pronunciado por una persona de renombre, los tonos de los discursos anteriores de la persona pueden codificarse en el estado oculto inicial. Otra técnica consiste en convertir el estado oculto inicial en un parámetro que se pueda entrenar. Aunque estas técnicas añaden pocos matices a la red, la inicialización del vector de estado oculto en ceros suele ser una opción eficaz.



### Funcionamiento de cada Unidad Recurrente: 
 

1. Tome como entrada el vector de estado previamente oculto y el vector de entrada actual.
Tenga en cuenta que, dado que el estado oculto y la entrada actual se tratan como vectores, cada elemento del vector se coloca en una dimensión diferente que es ortogonal a las otras dimensiones. Por lo tanto, cada elemento, cuando se multiplica por otro elemento, solo da un valor distinto de cero cuando los elementos involucrados son distintos de cero y los elementos están en la misma dimensión.

2. Elemento a elemento multiplica el vector de estado oculto por las ponderaciones de estado oculto y, de manera similar, realiza la multiplicación a elemento del vector de entrada actual y las ponderaciones de entrada actuales. Esto genera el vector de estado oculto parametrizado y el vector de entrada actual.
Tenga en cuenta que las ponderaciones de los diferentes vectores se almacenan en la matriz de ponderaciones entrenables.

3. Realice la suma vectorial de los dos vectores parametrizados y, a continuación, calcule la tangente hiperbólica elemental para generar el nuevo vector de estado oculto.

![](https://media.geeksforgeeks.org/wp-content/uploads/20190701130640/working1.png)

Durante el entrenamiento de la red recurrente, la red también genera una salida en cada paso de tiempo. Esta salida se utiliza para entrenar la red mediante el descenso de gradiente.

![](https://media.geeksforgeeks.org/wp-content/uploads/20190701141846/yt.png)

La retropropagación involucrada es similar a la utilizada en una red neuronal artificial típica con algunos cambios menores. Estos cambios se observan como:

Sea la salida prevista de la red en cualquier paso de tiempo $\overline{y_{t}}$     y la salida real sea $y_{t}$     . A continuación, el error en cada paso de tiempo viene dado por:

$$
E_{t} = -y_{t}log(\overline{y}_{t})     
$$

El error total viene dado por la suma de los errores en todos los periodos de tiempo.
$$
E = \sum _{t} E_{t}     
\rightarrow E = \sum _{t}-y_{t}log(\overline{y}_{t}) 
$$

Del mismo modo, el valor $\frac{\partial E}{\partial W}$ se puede calcular como la suma de gradientes en cada periodo de tiempo.

$$\frac{\partial E}{\partial W} = \sum _{t}\frac{\partial E_{t}}{\partial W}$$ 

Usando la regla de la cadena de cálculo y usando el hecho de que la salida en un paso de tiempo $t$ es una función del estado oculto actual de la unidad recurrente, surge la siguiente expresión:

$$\frac{\partial E_{t}}{\partial W} = \frac{\partial E_{t}}{\partial \overline{y}_{t}}\frac{\partial \overline{y}_{t}}{\partial h_{t}}\frac{\partial h_{t}}{\partial h_{t-1}}\frac{\partial h_{t-1}}{\partial h_{t-2}}......\frac{\partial h_{0}}{\partial W}$$     

Tenga en cuenta que la matriz de peso $W$ utilizada en la expresión anterior es diferente para el vector de entrada y el vector de estado oculto y solo se usa de esta manera por conveniencia notacional.

Así surge la siguiente expresión:

$$\frac{\partial E}{\partial W} = \sum _{t}\frac{\partial E_{t}}{\partial \overline{y}_{t}}\frac{\partial \overline{y}_{t}}{\partial h_{t}}\frac{\partial h_{t}}{\partial h_{t-1}}\frac{\partial h_{t-1}}{\partial h_{t-2}}......\frac{\partial h_{0}}{\partial W}$$ 

Por lo tanto, la retropropagación a través del tiempo solo difiere de una retropropagación típica en el hecho de que los errores en cada paso de tiempo se suman para calcular el error total.

![](https://media.geeksforgeeks.org/wp-content/uploads/20190701153452/back-prop.png)

Aunque la Red Neuronal Recurrente básica es bastante efectiva, puede sufrir un problema importante. En el caso de las redes profundas, el proceso de retropropagación puede dar lugar a los siguientes problemas:

* **Gradientes desvanecientes:** Esto ocurre cuando los gradientes se vuelven muy pequeños y tienden a cero.

* **Gradientes explosivos:** Esto ocurre cuando los gradientes se vuelven demasiado grandes debido a la retropropagación.

El problema de los gradientes explosivos se puede resolver mediante el uso de un truco: poniendo un umbral en los gradientes que se transmiten hacia atrás en el tiempo. Pero esta solución no se ve como una solución al problema y también puede reducir la eficiencia de la red. Para hacer frente a estos problemas, se desarrollaron dos variantes principales de las redes neuronales recurrentes: las redes de memoria a largo y corto plazo y las redes de unidades recurrentes cerradas.

Las redes neuronales recurrentes (RNN) son un tipo de red neuronal artificial diseñada para procesar datos secuenciales. A diferencia de las redes neuronales feedforward tradicionales, las RNN pueden tener en cuenta el estado anterior de la secuencia mientras procesan el estado actual, lo que les permite modelar dependencias temporales en los datos.

La característica clave de las RNN es la presencia de conexiones recurrentes entre las unidades ocultas, que permiten que la información pase de un paso de tiempo al siguiente. Esto significa que el estado oculto en cada unidad de tiempo no es solo una función de la entrada en esa unidad de tiempo, sino también una función del estado oculto anterior.

En una RNN, la entrada en cada periodo de tiempo suele ser un vector que representa el estado actual de la secuencia, y la salida en cada periodo de tiempo es un vector que representa el valor o la clasificación predichos en ese periodo de tiempo. El estado oculto también es un vector, que se actualiza en cada unidad de tiempo en función de la entrada actual y el estado oculto anterior.

La arquitectura básica de RNN sufre el problema del gradiente de desaparición, lo que puede dificultar el entrenamiento en secuencias largas. Para abordar este problema, se han desarrollado varias variantes de RNN, como las redes de memoria a corto plazo (LSTM) y de unidad recurrente cerrada (GRU), que utilizan puertas especializadas para controlar el flujo de información a través de la red y abordar el problema del gradiente de desaparición.

Las aplicaciones de las RNN incluyen el reconocimiento de voz, el modelado del lenguaje, la traducción automática, el análisis de sentimientos y la predicción de acciones, entre otras. En general, las RNN son una herramienta eficaz para procesar datos secuenciales y modelar dependencias temporales, lo que las convierte en un componente importante de muchas aplicaciones de aprendizaje automático.
 

### Las ventajas de las Redes Neuronales Recurrentes (RNN) son:

* Capacidad para procesar datos secuenciales: las RNN pueden procesar datos secuenciales de diferentes longitudes, lo que las hace útiles en aplicaciones como el procesamiento del lenguaje natural, el reconocimiento de voz y el análisis de series temporales.

* Memoria: Las RNN tienen la capacidad de retener información sobre las entradas anteriores en la secuencia mediante el uso de estados ocultos. Esto permite a las RNN realizar tareas como predecir la siguiente palabra en una oración o pronosticar los precios de las acciones.

* Versatilidad: Las RNN se pueden utilizar para una amplia variedad de tareas, incluida la clasificación, la regresión y el mapeo de secuencia a secuencia.

* Flexibilidad: las RNN se pueden combinar con otras arquitecturas de redes neuronales, como las redes neuronales convolucionales (CNN) o las redes neuronales feedforward, para crear modelos híbridos para tareas específicas.

### Sin embargo, también hay algunas desventajas de las RNN:

* Problema del gradiente de fuga: El problema del gradiente de fuga puede ocurrir en RNN, particularmente en aquellas con muchas capas o secuencias largas, lo que dificulta el aprendizaje de dependencias a largo plazo.

* Costoso desde el punto de vista computacional: Las RNN pueden ser costosas desde el punto de vista computacional, especialmente cuando se procesan secuencias largas o se utilizan arquitecturas complejas.

* Falta de interpretabilidad: Las RNN pueden ser difíciles de interpretar, particularmente en términos de comprender cómo la red está haciendo predicciones o decisiones.

* En general, aunque las RNN tienen algunas desventajas, su capacidad para procesar datos secuenciales y retener la memoria de las entradas anteriores las convierte en una herramienta eficaz para muchas aplicaciones de aprendizaje automático.


```python
import tensorflow as tf
import numpy as np

# Crear datos de ejemplo
num_samples = 1000
sequence_length = 5

X = np.random.random((num_samples, sequence_length, 1))  # Datos de entrada: 1000 secuencias, cada una de longitud 5
y = np.mean(X, axis=1) > 0.5  # Etiquetas: Verdadero si la media de la secuencia es mayor que 0.5

# Crear el modelo RNN
model = tf.keras.models.Sequential([
    tf.keras.layers.SimpleRNN(50, input_shape=(sequence_length, 1)),  # Capa RNN con 50 unidades
    tf.keras.layers.Dense(1, activation='sigmoid')  # Capa de salida con activación sigmoide para clasificación binaria
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(X, y, epochs=10, batch_size=32)

# Evaluar el modelo
loss, accuracy = model.evaluate(X, y)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

El codigo anterior realiza lo siguiente:

* Genera datos de ejemplo: Crea secuencias aleatorias de números y las clasifica según su media.
* Construye el modelo: Utiliza tf.keras para crear una RNN simple con una capa SimpleRNN y una capa de salida Dense.
* Compila el modelo: Define el optimizador, la función de pérdida y las métricas.
* Entrena el modelo: Usa los datos generados para entrenar el modelo.
* Evalúa el modelo: Muestra la pérdida y la precisión del modelo en los mismos datos.


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
num_samples = 1000
sequence_length = 5

X = np.random.random((num_samples, sequence_length, 1))
y = np.mean(X, axis=1) > 0.5

# Crear el modelo RNN
model = tf.keras.models.Sequential([
    tf.keras.layers.SimpleRNN(50, input_shape=(sequence_length, 1)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo y guardar el historial
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)

# Graficar la precisión y la pérdida
plt.figure(figsize=(12, 4))

# Precisión
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Pérdida
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

Este código realiza las siguientes acciones adicionales:

* Guarda el historial de entrenamiento: Al entrenar el modelo con `model.fit()`, guardamos el resultado en una variable `history`. Esta variable contiene información sobre la precisión y la pérdida durante el entrenamiento.
* Grafica la precisión y la pérdida: Usamos matplotlib para crear gráficas de la precisión y la pérdida a lo largo de las épocas. También incluimos las métricas para el conjunto de validación si decides usar una división de validación.

## Transformers

Los "transformers" en redes neuronales son un tipo de arquitectura de red neuronal que ha revolucionado el campo del procesamiento del lenguaje natural (NLP, por sus siglas en inglés). A diferencia de las arquitecturas anteriores, como las redes neuronales recurrentes (RNN) y las redes de memoria a corto plazo largo (LSTM), los transformers han demostrado ser más eficientes y efectivos para manejar secuencias de datos, especialmente texto.

Los transformers se basan en un mecanismo denominado "atención", que permite a la red enfocarse en diferentes partes de la secuencia de entrada al mismo tiempo. Esto es crucial para entender el contexto y las relaciones complejas en el lenguaje.

Principales características de los transformers:

**Mecanismo de Atención:** La atención permite al modelo ponderar la importancia relativa de diferentes palabras en una oración. Esto ayuda al modelo a entender mejor el contexto y las relaciones entre palabras, lo que es crucial para tareas como la traducción automática y la comprensión del lenguaje.

**Paralelización:** A diferencia de las RNN y las LSTM, que procesan los datos secuencialmente, los transformers pueden procesar toda la secuencia de entrada de una vez, lo que permite un procesamiento paralelo y una velocidad de entrenamiento significativamente mayor.

**Capas Codificadoras y Decodificadoras:** En tareas como la traducción automática, los transformers utilizan capas de codificación para procesar la entrada (por ejemplo, un texto en un idioma) y capas de decodificación para generar la salida (el texto traducido en otro idioma).

**Autoatención:** Es una forma de atención que permite a cada posición en la secuencia de entrada atender a todas las posiciones en la misma secuencia. Esto mejora la capacidad del modelo para captar las dependencias a largo plazo.

**Flexibilidad y Generalización:** Los transformers se han utilizado con éxito en una amplia gama de tareas de NLP, demostrando una gran capacidad de generalización y adaptabilidad.

El modelo "Transformer" fue introducido en el artículo "Attention is All You Need" de Vaswani et al. en 2017 y ha sido la base para el desarrollo de modelos de lenguaje muy conocidos y avanzados, como BERT, GPT, T5 y muchos otros, que han establecido nuevos estándares en diversas tareas de procesamiento de lenguaje.
