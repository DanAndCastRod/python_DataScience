# Técnicas Heurísticas y metaheurísticas

Las técnicas heurísticas y metaheurísticas son métodos utilizados para encontrar soluciones aceptables para problemas complejos, especialmente en casos donde los métodos exactos son impracticables debido al tiempo o recursos computacionales requeridos. Estas técnicas son fundamentales en la optimización y la resolución de problemas en diversas áreas como la inteligencia artificial, ingeniería, economía y logística.

### Heurísticas

1. **Definición**: Las heurísticas son estrategias basadas en la experiencia, intuición o reglas prácticas, diseñadas para encontrar soluciones suficientemente buenas para un problema, especialmente cuando un enfoque completo no es posible o práctico.

2. **Características**:
   - **Eficiencia**: Suelen ser más rápidas que los métodos exactos.
   - **Soluciones Aproximadas**: No garantizan la mejor solución, pero a menudo son lo suficientemente buenas.
   - **Especificidad**: Generalmente son específicas para un problema o clase de problemas.

3. **Ejemplos**:
   - Algoritmo de búsqueda local.
   - Algoritmo voraz (greedy).
   - Algoritmo de eliminación de ramas.

### Metaheurísticas

1. **Definición**: Las metaheurísticas son un nivel más alto de estrategias heurísticas diseñadas para guiar un proceso de búsqueda subyacente y explorar el espacio de soluciones de manera eficiente. Son más generales y flexibles que las heurísticas simples.

2. **Características**:
   - **Generalidad**: Pueden aplicarse a una amplia variedad de problemas.
   - **Balance**: Buscan un equilibrio entre exploración (buscar en nuevas áreas) y explotación (profundizar en áreas conocidas).
   - **Iterativas**: Mejoran progresivamente las soluciones a través de iteraciones.

3. **Ejemplos**:
   - Algoritmos genéticos.
   - Recocido simulado (Simulated Annealing).
   - Búsqueda tabú.
   - Optimización por enjambre de partículas (Particle Swarm Optimization).

### Diferencias Clave

- **Especificidad vs. Generalidad**: Las heurísticas suelen ser específicas para un problema, mientras que las metaheurísticas son marcos más generales aplicables a diversos problemas.
- **Complejidad**: Las metaheurísticas son típicamente más complejas y abarcan un rango más amplio de técnicas de búsqueda que las heurísticas simples.

Ambas son herramientas valiosas en campos donde los problemas son demasiado complejos para ser resueltos por métodos exactos, ofreciendo un equilibrio entre tiempo de cálculo y calidad de la solución obtenida.

## Desarrollo

Vamos a explorar el desarrollo conceptual y teórico, incluyendo aspectos matemáticos, de tres técnicas heurísticas populares: Algoritmos Voraces (Greedy Algorithms), Búsqueda Local, Heurísticas Constructivas.

### 1. Algoritmos Voraces (Greedy Algorithms)

**Concepto**: Los algoritmos voraces toman decisiones secuenciales basadas en la mejor opción local en cada paso, con la esperanza de encontrar una solución global óptima.

**Desarrollo Matemático**:

- En cada paso, se elige la opción que parece ser la mejor en ese momento.

- No se reconsideran las decisiones tomadas.

- Ejemplo: En el problema del cambio de monedas, si se tienen denominaciones de 1, 5, 10, y se necesita dar cambio para 16, el algoritmo escogerá una moneda de 10, luego una de 5, y finalmente una de 1.

**Teoría**:

- Eficiencia: O(n) en muchos casos.

- No siempre garantiza la solución óptima.

- Ejemplo de aplicación: Algoritmo de Kruskal para encontrar el árbol de expansión mínima en grafos.

### 2. Búsqueda Local

**Concepto**: Partiendo de una solución inicial, la búsqueda local explora las soluciones vecinas para encontrar una mejor.

**Desarrollo Matemático**:

- Se define una "vecindad" para cada solución.

- Se explora esta vecindad para encontrar una solución mejor.

- Ejemplo: En un problema de optimización, si se tiene una solución $x$, se explora alrededor de $x$ (por ejemplo, $x + \Delta$ y $x
- \Delta$).

**Teoría**:

- A menudo se queda en óptimos locales.

- Variaciones incluyen: Hill Climbing, Simulated Annealing.

- Ejemplo de aplicación: Optimización de rutas en problemas de viajante de comercio (TSP).

### 3. Heurísticas Constructivas

**Concepto**: Construyen una solución desde cero, agregando componentes paso a paso hasta completar la solución.

**Desarrollo Matemático**:

- Se comienza con un conjunto vacío o parcialmente lleno.

- En cada paso, se agrega el componente que parece ser el mejor según algún criterio.

- Ejemplo: En un problema de programación de horarios, se comienza sin ningún horario asignado y se van llenando los espacios uno a uno.

**Teoría**:

- Se utiliza una función de evaluación para seleccionar el componente a agregar.

- Es una estrategia general que se puede adaptar a muchos problemas.

Estas técnicas heurísticas se basan en principios matemáticos y lógicos para abordar problemas complejos de optimización, ofreciendo un equilibrio entre el tiempo de cálculo y la calidad de las soluciones obtenidas.

### Ejemplos
Para ilustrar las tres técnicas heurísticas mencionadas - Algoritmos Voraces, Búsqueda Local, Heurísticas Constructivas- vamos a utilizar un contexto común: el problema del Viajante de Comercio (Traveling Salesman Problem, TSP). Este es un problema clásico de optimización donde se busca la ruta más corta que pasa por un conjunto de ciudades, visitando cada una una sola vez y regresando al punto de inicio.

1. **Algoritmos Voraces**: Seleccionaremos la ciudad más cercana no visitada en cada paso.
2. **Búsqueda Local**: Partiremos de una ruta y haremos pequeñas modificaciones (como intercambiar dos ciudades) para mejorar la distancia total.
3. **Heurísticas Constructivas**: Construiremos la ruta agregando ciudades de manera que se minimice el incremento de la distancia total en cada paso.

Para estos ejemplos, utilizaremos un conjunto de ciudades con coordenadas generadas aleatoriamente. Visualizaremos cada técnica con gráficos para entender mejor cómo funcionan.

### Preparación del Entorno y Datos

Primero, prepararemos el entorno de Python y generaremos las coordenadas de las ciudades.



```python
# @title Preparación de data
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
import random

# Generar coordenadas aleatorias para las ciudades
np.random.seed(0)  # Para reproducibilidad
num_cities = 13 #@param
coordinates = {i: (np.random.rand(), np.random.rand()) for i in range(num_cities)}

# Función para visualizar la ruta
def plot_route(route, coordinates):
    plt.figure(figsize=(8, 6))
    x = [coordinates[i][0] for i in route]
    y = [coordinates[i][1] for i in route]
    plt.plot(x, y, 'o-', color='blue')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Traveling Salesman Route')
    plt.show()

# Función para calcular la distancia total de la ruta
def route_distance(route, coordinates):
    total_distance = 0
    for i in range(len(route)):
        total_distance += np.linalg.norm(np.array(coordinates[route[i]]) - np.array(coordinates[route[(i + 1) % len(route)]]))
    return total_distance
y = [coordinates[i][1] for i in coordinates]
x = [coordinates[i][0] for i in coordinates]
plt.scatter(x,y)
xy = [(x[i],y[i]) for i in range(len(x))]
for i in coordinates.keys():
  plt.annotate(i,(x[i]*1.01,y[i]*1.01))
```


Ahora, procederemos con la implementación de cada técnica heurística para el problema del Viajante de Comercio (TSP) y visualizaremos la ruta resultante para cada técnica.

### 1. Algoritmos Voraces

Implementaremos un algoritmo voraz seleccionando la ciudad más cercana no visitada en cada paso.



```python
# @markdown
def greedy_tsp(coordinates):
    current_city = list(coordinates.keys())[0]
    route = [current_city]
    while len(route) < len(coordinates):
        next_city = min({c: np.linalg.norm(np.array(coordinates[current_city]) - np.array(coordinates[c]))
                         for c in coordinates if c not in route}, key=lambda c: coordinates[c])
        route.append(next_city)
        current_city = next_city
    return route

# Aplicando el algoritmo voraz y visualizando la ruta
greedy_route = greedy_tsp(coordinates)
plot_route(greedy_route, coordinates)
```

### 2. Búsqueda Local
Implementaremos un algoritmo de búsqueda local intercambiando dos ciudades y manteniendo el cambio si mejora la distancia total.

```python
# @markdown
def local_search_tsp(coordinates):
    route = list(coordinates.keys())
    best_distance = route_distance(route, coordinates)
    improved = True
    while improved:
        improved = False
        for i in range(len(route)):
            for j in range(i+1, len(route)):
                new_route = route[:]
                new_route[i], new_route[j] = new_route[j], new_route[i]
                new_distance = route_distance(new_route, coordinates)
                if new_distance < best_distance:
                    route = new_route
                    best_distance = new_distance
                    improved = True
    return route

# Aplicando búsqueda local y visualizando la ruta
local_search_route = local_search_tsp(coordinates)
plot_route(local_search_route, coordinates)
```

### 3. Heurísticas Constructivas
Construiremos la ruta agregando la ciudad que genere el menor incremento en la distancia total.

```python
# @markdown
def constructive_tsp(coordinates):
    remaining_cities = list(coordinates.keys())
    current_city = remaining_cities.pop(0)
    route = [current_city]
    while remaining_cities:
        next_city = min(remaining_cities, key=lambda c: np.linalg.norm(np.array(coordinates[current_city]) - np.array(coordinates[c])))
        route.append(next_city)
        current_city = next_city
        remaining_cities.remove(next_city)
    return route

# Aplicando heurística constructiva y visualizando la ruta
constructive_route = constructive_tsp(coordinates)
plot_route(constructive_route, coordinates)
```

### Algoritmo Genético


El Algoritmo Genético (AG) es una metaheurística inspirada en el proceso de evolución natural y pertenece a la clase de los algoritmos evolutivos. Su desarrollo implica los siguientes conceptos y pasos:

1. **Población Inicial**: El AG comienza con un conjunto de soluciones (denominado población), donde cada solución (individuo) está generalmente representada por una cadena de caracteres, a menudo una cadena de bits.

2. **Aptitud**: Cada individuo en la población tiene una función de aptitud que determina qué tan "buena" es la solución que representa. La aptitud suele estar relacionada con el objetivo del problema.

3. **Selección**: Se seleccionan individuos para la reproducción. Los individuos con mayor aptitud tienen más probabilidades de ser seleccionados.

4. **Cruce (Reproducción)**: Pares de individuos (padres) se cruzan para producir descendientes (hijos), combinando partes de las cadenas de caracteres de los padres.

5. **Mutación**: Con cierta probabilidad, las cadenas de caracteres de los individuos pueden sufrir cambios aleatorios. Esto ayuda a mantener la diversidad genética en la población y permite explorar nuevas soluciones.

6. **Reemplazo**: Los hijos reemplazan a algunos de los padres en la población, y el proceso se repite.

El AG itera sobre estos pasos hasta que se cumple un criterio de parada, como alcanzar un número máximo de generaciones o encontrar una solución satisfactoria.

### Ejemplo en Python

Vamos a implementar un AG simple para optimizar una función objetivo. En este ejemplo, nuestro objetivo será encontrar el máximo de la función \( f(x) = x^2 \) en el intervalo \([0, 31]\), utilizando una representación binaria para los números.

Pasos para la implementación:

1. **Representación de Individuos**: Cada número en el intervalo \([0, 31]\) se representa como un string binario de 5 bits.

2. **Función de Aptitud**: Será el valor de \( f(x) = x^2 \).

3. **Selección, Cruce y Mutación**: Implementaremos funciones básicas para estos procesos.

4. **Visualización**: Usaremos gráficos para mostrar la evolución de la aptitud a lo largo de las generaciones.

Vamos a implementar esto en Python con visualización gráfica.

La gráfica muestra cómo la aptitud promedio de la población evoluciona a lo largo de las generaciones. En este caso, la aptitud se refiere al valor de \( f(x) = x^2 \), donde \( x \) es un número representado por una cadena de bits en cada individuo de la población.

El algoritmo inicia con una población aleatoria y, a través de los procesos de selección, cruce y mutación, y la supervivencia de los más aptos, la aptitud promedio de la población tiende a aumentar con cada generación. Esto sugiere que el algoritmo está convergiendo gradualmente hacia una solución óptima, es decir, una representación binaria que corresponda al máximo valor de \( f(x) \) en el rango definido.

En resumen, este ejemplo ilustra cómo un Algoritmo Genético puede ser utilizado para encontrar soluciones aproximadas a problemas de optimización, aprovechando los principios de la evolución natural como la selección y la mutación. La visualización gráfica proporciona una perspectiva clara de cómo la aptitud de la población evoluciona y mejora a lo largo del tiempo.

```python
import random
import matplotlib.pyplot as plt
import math

def individual(length, min, max):
    """Crear un individuo como una cadena de bits."""
    return [random.randint(min, max) for _ in range(length)]

def population(count, length, min, max):
    """Crear una población de individuos."""
    return [individual(length, min, max) for _ in range(count)]

def fitness(individual):
    """Función de aptitud: evaluar qué tan bueno es un individuo."""
    number = int("".join(map(str, individual)), 2)
    # return (number)**2
    # return -(((number-15)/(2.5)))**(2)+32
    return math.sin(number*math.pi/360)

def grade(pop):
    """Encontrar la aptitud promedio de una población."""
    summed = sum(fitness(x) for x in pop)
    return summed / (len(pop) * 1.0)

def evolve(pop, retain=0.2, random_select=0.05, mutate=0.01):
    graded = [(fitness(x), x) for x in pop]
    graded = [x[1] for x in sorted(graded, reverse=True)]
    retain_length = int(len(graded)*retain)
    parents = graded[:retain_length]

    # Selección aleatoria para mantener la diversidad
    for individual in graded[retain_length:]:
        if random_select > random.random():
            parents.append(individual)

    # Mutación
    for individual in parents:
        if mutate > random.random():
            pos_to_mutate = random.randint(0, len(individual)-1)
            individual[pos_to_mutate] = random.randint(min(individual), max(individual))

    # Cruce
    parents_length = len(parents)
    desired_length = len(pop) - parents_length
    children = []
    while len(children) < desired_length:
        male = random.randint(0, parents_length-1)
        female = random.randint(0, parents_length-1)
        if male != female:
            male = parents[male]
            female = parents[female]
            half = len(male) // 2
            child = male[:half] + female[half:]
            children.append(child)

    parents.extend(children)
    return parents

# Parámetros
# Número de individuos en la población
p_count = 100  #@param
 # Longitud de cada individuo (5 bits)
i_length = 5  #@param
# Mínimo valor del bit
i_min = 0      #@param
# Máximo valor del bit
i_max = 1      #@param
# Número de generaciones
generations = 100  #@param

# Crear la población inicial
p = population(p_count, i_length, i_min, i_max)

# Evolucionar la población
fitness_history = [grade(p),]
for i in range(generations):
    p = evolve(p)
    fitness_history.append(grade(p))

# Visualización
plt.plot(range(generations + 1), fitness_history)
plt.title("Desarrollo de la Aptitud a lo Largo de las Generaciones")
plt.xlabel("Generaciones")
plt.ylabel("Aptitud Promedio")
plt.show()
```

```python
import time
from IPython.display import clear_output

# Actualizamos la función evolve para que devuelva también la aptitud máxima de cada generación
def evolve_with_max_fitness(pop, retain=0.2, random_select=0.05, mutate=0.01):
    graded = [(fitness(x), x) for x in pop]
    graded = sorted(graded, reverse=True)
    max_fitness = graded[0][0]  # Máxima aptitud de esta generación
    graded = [x[1] for x in graded]
    retain_length = int(len(graded)*retain)
    parents = graded[:retain_length]

    # Selección aleatoria y mutación igual que antes

    # Selección aleatoria para mantener la diversidad
    for individual in graded[retain_length:]:
        if random_select > random.random():
            parents.append(individual)

    # Mutación
    for individual in parents:
        if mutate > random.random():
            pos_to_mutate = random.randint(0, len(individual)-1)
            individual[pos_to_mutate] = random.randint(min(individual), max(individual))

    # Cruce
    parents_length = len(parents)
    desired_length = len(pop) - parents_length
    children = []
    while len(children) < desired_length:
        male = random.randint(0, parents_length-1)
        female = random.randint(0, parents_length-1)
        if male != female:
            male = parents[male]
            female = parents[female]
            half = len(male) // 2
            child = male[:half] + female[half:]
            children.append(child)

    parents.extend(children)
    return parents, max_fitness

# Parámetros
# Reducimos el número de generaciones para la demostración
generations = 10 #@param
# Crear la población inicial
p = population(p_count, i_length, i_min, i_max)

# Evolucionar la población y visualizar
for i in range(generations):
    p, max_fitness = evolve_with_max_fitness(p)
    clear_output(wait=True)
    plt.figure(figsize=(10, 6))
    plt.title(f"Generación: {i+1}, Máxima Aptitud: {max_fitness}")
    plt.xlabel("Individuo")
    plt.ylabel("Aptitud")
    plt.bar(range(len(p)), [fitness(x) for x in p], color='blue')
    plt.show()
    time.sleep(0.5)  # Pausa de 0.5 segundos entre generaciones

```

### Recocido Simulado

El Recocido Simulado (Simulated Annealing, SA) es una metaheurística inspirada en el proceso físico de calentamiento y enfriamiento controlado de materiales para aumentar la resistencia de los mismos. En el contexto de la optimización, se utiliza para encontrar una buena aproximación a la solución óptima de problemas complejos.

### Conceptualización del Recocido Simulado

1. **Inspiración Física**: Se basa en el proceso de recocido en metalurgia. El material se calienta y luego se enfría lentamente para que sus partículas se reorganicen gradualmente, minimizando su energía interna.

2. **Analogía en Optimización**:
   - **Estado**: Representa una solución posible del problema.
   - **Energía**: Es equivalente a la función objetivo del problema. Se busca minimizar esta "energía".
   - **Temperatura**: Controla la probabilidad de aceptar soluciones peores. Alta temperatura permite explorar más el espacio de soluciones, evitando mínimos locales.

3. **Proceso**:
   - **Inicialización**: Se comienza con una solución inicial y una temperatura alta.
   - **Iteraciones**: En cada paso, se genera una nueva solución cercana a la actual (vecindad).
   - **Aceptación**: La nueva solución se acepta si es mejor, o con cierta probabilidad si es peor, dependiendo de la temperatura.
   - **Enfriamiento**: La temperatura se reduce gradualmente, disminuyendo la probabilidad de aceptar peores soluciones.

4. **Enfriamiento**: La forma en que se reduce la temperatura es crucial. Puede ser lineal, exponencial, etc.

5. **Criterio de Parada**: El algoritmo termina después de un número fijo de iteraciones, o cuando la temperatura es suficientemente baja.

### Ejemplo en Python

Implementaremos un ejemplo básico de Recocido Simulado para encontrar el mínimo de una función matemática sencilla. Utilizaremos gráficos para visualizar el progreso del algoritmo.

**Función Objetivo**: $ f(x) = x^2 $

**Meta**: Encontrar el mínimo de $ f $ en el rango $[-10, 10]$.

Vamos a programar esto y visualizar el proceso.



```python
#@title Ejemplo
import numpy as np
import matplotlib.pyplot as plt

# Función objetivo
def objective_function(x):
    return x ** 2

# Genera un vecino cercano
def neighbor(current_state):
    step = np.random.uniform(-1, 1)
    new_state = current_state + step
    return np.clip(new_state, -10, 10)  # Mantener dentro de los límites

# Función de aceptación
def acceptance_probability(old_cost, new_cost, temperature):
    if new_cost < old_cost:
        return 1.0
    else:
        return np.exp((old_cost - new_cost) / temperature)

# Algoritmo de recocido simulado
def simulated_annealing(objective, initial_state, initial_temp, final_temp, alpha):
    current_state = initial_state
    current_temp = initial_temp
    states, costs = [], []

    while current_temp > final_temp:
        new_state = neighbor(current_state)
        new_cost = objective(new_state)
        old_cost = objective(current_state)

        # Decidir si se acepta el nuevo estado
        if acceptance_probability(old_cost, new_cost, current_temp) > np.random.rand():
            current_state = new_state

        states.append(current_state)
        costs.append(new_cost)
        current_temp *= alpha  # Enfriamiento

    return states, costs

# Parámetros del algoritmo
initial_temp = 10.0
final_temp = 0.1
alpha = 0.9
initial_state = np.random.uniform(-10, 10)

# Ejecutar el algoritmo
states, costs = simulated_annealing(objective_function, initial_state, initial_temp, final_temp, alpha)

# Gráfico de la evolución de la solución
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(states)
plt.title("Evolución del Estado")
plt.xlabel("Iteración")
plt.ylabel("Estado (x)")

plt.subplot(1, 2, 2)
plt.plot(costs)
plt.title("Evolución del Costo")
plt.xlabel("Iteración")
plt.ylabel("Costo (f(x))")

plt.tight_layout()
plt.show()
```

El gráfico muestra cómo el algoritmo de Recocido Simulado evoluciona a lo largo de las iteraciones para el problema de minimizar la función $ f(x) = x^2 $ en el rango de $[-10, 10]$.

- **Evolución del Estado**: Representa los valores de $ x $ considerados en cada iteración. Inicialmente, el algoritmo explora una amplia gama de valores debido a la alta temperatura, lo que permite saltar incluso a valores peores. A medida que la temperatura disminuye, el algoritmo comienza a enfocarse más en el área alrededor del mínimo local encontrado.

- **Evolución del Costo**: Muestra el valor de la función objetivo $ f(x) $ en cada iteración. Aunque hay fluctuaciones al principio (debido a la aceptación de peores soluciones), gradualmente el costo disminuye y se estabiliza a medida que la temperatura se reduce.

### Puntos Clave

1. **Exploración y Explotación**: El Recocido Simulado equilibra estos dos aspectos. La alta temperatura inicial permite explorar el espacio de soluciones (evitando mínimos locales prematuros), y la disminución gradual enfoca la búsqueda hacia la explotación del mejor mínimo encontrado.

2. **Parámetros Clave**: La temperatura inicial, el factor de enfriamiento (alpha), y el criterio de parada son fundamentales para el rendimiento del algoritmo.

3. **Aplicabilidad**: Aunque este es un ejemplo simple, el Recocido Simulado es aplicable a una amplia gama de problemas de optimización, incluyendo aquellos con espacios de solución de alta dimensión y no lineales.

El Recocido Simulado es una herramienta poderosa en la optimización, especialmente útil en problemas donde los métodos tradicionales pueden quedar atrapados en mínimos locales.

```python
#@title iter
import time
from IPython.display import clear_output

# Función para mostrar la evolución del algoritmo de manera iterativa
def iterative_simulated_annealing_display(objective, initial_state, initial_temp, final_temp, alpha):
    current_state = initial_state
    current_temp = initial_temp
    states, costs = [initial_state], [objective(initial_state)]

    # Configuración del gráfico
    plt.figure(figsize=(14, 6))

    while current_temp > final_temp:
        new_state = neighbor(current_state)
        new_cost = objective(new_state)
        old_cost = objective(current_state)

        # Decidir si se acepta el nuevo estado
        if acceptance_probability(old_cost, new_cost, current_temp) > np.random.rand():
            current_state = new_state

        states.append(current_state)
        costs.append(new_cost)
        current_temp *= alpha  # Enfriamiento

        # Actualizar gráficos
        clear_output(wait=True)
        plt.subplot(1, 2, 1)
        plt.plot(states, 'b.-')
        plt.title(f"Evolución del Estado (Temp = {current_temp:.2f})")
        plt.xlabel("Iteración")
        plt.ylabel("Estado (x)")
        plt.ylim(-10, 10)

        plt.subplot(1, 2, 2)
        plt.plot(costs, 'r.-')
        plt.title("Evolución del Costo")
        plt.xlabel("Iteración")
        plt.ylabel("Costo (f(x))")
        plt.ylim(0, max(costs) + 5)

        plt.tight_layout()
        plt.show()
        time.sleep(0.5)

    return states, costs

# Ejecución del algoritmo con visualización iterativa
iterative_simulated_annealing_display(objective_function, initial_state, initial_temp, final_temp, alpha)
```

```python
#@title iter
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

# Reutilizando las funciones definidas anteriormente
def objective_function(x):
    return x ** 2

def neighbor(current_state):
    step = np.random.uniform(-1, 1)
    new_state = current_state + step
    return np.clip(new_state, -10, 10)

def acceptance_probability(old_cost, new_cost, temperature):
    if new_cost < old_cost:
        return 1.0
    else:
        return np.exp((old_cost - new_cost) / temperature)

# Nuevo algoritmo de recocido simulado que actualiza un mapa de calor
def simulated_annealing_heatmap(objective, initial_state, initial_temp, final_temp, alpha, num_bins):
    current_state = initial_state
    current_temp = initial_temp
    heatmap, bin_edges = np.histogram([], bins=num_bins, range=(-10, 10))
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

    while current_temp > final_temp:
        new_state = neighbor(current_state)
        new_cost = objective(new_state)
        old_cost = objective(current_state)

        if acceptance_probability(old_cost, new_cost, current_temp) > np.random.rand():
            current_state = new_state

        # Actualizar el mapa de calor
        state_index = np.digitize(current_state, bin_edges) - 1
        heatmap[state_index] += 1

        # Actualizar la visualización
        clear_output(wait=True)
        plt.bar(bin_centers, heatmap, width=bin_edges[1] - bin_edges[0])
        plt.title(f"Mapa de Calor del Recocido Simulado (Temp = {current_temp:.2f})")
        plt.xlabel("Estado (x)")
        plt.ylabel("Frecuencia")
        plt.show()

        time.sleep(0.5)
        current_temp *= alpha

    return heatmap

# Parámetros
initial_temp = 10.0
final_temp = 0.1
alpha = 0.9
initial_state = np.random.uniform(-10, 10)
num_bins = 20

# Ejecutar el algoritmo
simulated_annealing_heatmap(objective_function, initial_state, initial_temp, final_temp, alpha, num_bins)
```
